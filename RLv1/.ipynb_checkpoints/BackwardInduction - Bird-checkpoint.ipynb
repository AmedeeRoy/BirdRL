## Export Gif

```python
from rlbirdv1 import *

x = [i for i in range(3)]
y = [i for i in range(3)]
y.reverse()

# List of coordinates for islands
island = [(0, 0)]
birdStart = (0, 0)

TILESIZE = 40
SCREEN_WIDTH = TILESIZE*(len(x)+3)
SCREEN_HEIGHT = TILESIZE * len(y)

reward = {'lose' : -1000,
          'win' : 1000}

actions = {
    0 : K_LEFT,
    1 : K_RIGHT,
    2 : K_DOWN,
    3 : K_UP,
    4 : K_s,
    5 : K_d
}

N = 20

# load game run.py
game = RLBird(width = SCREEN_WIDTH, height = SCREEN_HEIGHT, x = x, y = y,\
              init_bird_position = birdStart, island_position = island, \
              energyMax = 10, catchMax = 2, costMove = -1, costDive = -1, gainFish = 2, factorFishFly = 0.75,\
              nbStep = N, reward = reward)
```

```python
from ple import PLE
import mdptoolbox

p = PLE(game, fps=1000, frame_skip=1, num_steps=1, force_fps=True, display_screen=True, reward_values = reward)
p.init()

# compute optimal policy
print('Initializing...')
P = game.getTransitionProbabilities()
R = game.getRewardMatrix()
h = game.getTerminalRewardMatrix()
print('Backward Induction...')
backWard = mdptoolbox.mdp.FiniteHorizon(P, R, 1, N, h = h)
backWard.run()
```

```python
saveFishMap = game.map.fishMap.copy()
```

```python
import imageio
images = []
dt = 0
while(not p.game_over()):
    s = game.getGameState()
    idx = game.listStates.state2idx(s)
    a = backWard.policy[idx, dt]
    reward = p.act(actions[int(a)])
    
    pic = './results/screenshot'+str(dt)+'.jpeg'
    pygame.image.save(game.screen, pic)
    images.append(imageio.imread(pic))

    dt +=1
    
imageio.mimsave('./results/movie.gif', images)
pygame.quit()
```

## Run several time the same map

```python
nb_games = 20
cumulated = np.zeros((nb_games))
for i in range(nb_games):
    dt = 0
    game.init()
    game.updateFishMap(saveFishMap.copy())
    
    while(not game.game_over()):      
        s = game.getGameState()
        idx = game.listStates.state2idx(s)
        a = backWard.policy[idx, dt]
        if a==0:
            game.bird.moveLeft(game.map)
        if a==1:
            game.bird.moveRight(game.map)
        if a==2:
            game.bird.moveDown(game.map)
        if a==3:
            game.bird.moveUp(game.map)
        if a==4:
            game.bird.stay()
        if a==5:
            game.bird.dive(game.map, factorFishFly = 0.75)
        
        dt += 1 
    
    cumulated[i] = game.getScore()
cumulated
```
